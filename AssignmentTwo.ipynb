```
FUNCTION LinearSearch(list, target):
    FOR each index i from 0 to length(list) - 1:
        IF list[i] is equal to target:
            RETURN i
    RETURN -1
```

```
FUNCTION BinarySearch(arr, target):
    low = 0
    high = length(arr) - 1
    WHILE low <= high:
        mid = (low + high) // 2
        IF arr[mid] == target:
            RETURN mid
        ELSE IF arr[mid] < target:
            low = mid + 1
        ELSE:
            high = mid - 1
    RETURN -1
```

````python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# Testing
test_list = [3, 7, 12, 19, 21, 25, 30, 38, 42, 50, 55, 61, 70, 88, 99]
target = 42

print(f"Linear Search Index: {linear_search(test_list, target)}")
print(f"Binary Search Index: {binary_search(test_list, target)}")
````

````python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
````

Difference Between sort() and sorted()

Sort() : This is a list method that sorts the list in place. It modifies the original list and returns none.

Sorted() : This is a built in function that returns a new sorted list from the elements of any iterable. The orignal data remains unchanged.

````python
nums = [45, 12, 89, 7, 33]

# 1. Ascending
nums.sort()
print(f"Ascending: {nums}")

# 2. Descending
nums.sort(reverse=True)
print(f"Descending: {nums}")

# 3. List of tuples by second value
tuple_list = [(1, 'banana'), (2, 'apple'), (3, 'cherry')]
sorted_tuples = sorted(tuple_list, key=lambda x: x[1])
print(f"Sorted Tuples: {sorted_tuples}")
````

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node

    def delete_node(self, key):
        temp = self.head
        if temp and temp.data == key:
            self.head = temp.next
            return
        prev = None
        while temp and temp.data != key:
            prev = temp
            temp = temp.next
        if not temp: return
        prev.next = temp.next

    def display(self):
        curr = self.head
        while curr:
            print(curr.data, end=" -> ")
            curr = curr.next
        print("None")
````

Differentiate Types

Singly Linked List : Each node points to the next node only.

Doubly Linked List : Each node points to both the next node and the previous node.

Circular Linked List : The last node points back to the first node instead of None.

Advantages of Array

````python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, value):
        self.stack.append(value)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        return "Stack is empty"

    def is_empty(self):
        return len(self.stack) == 0

# ii. Operations
my_stack = Stack()
my_stack.push(20)
my_stack.push(50)
print(f"Popped value: {my_stack.pop()}") # Should return 50
````

Differences

Normal Queue : Follows FIFO(First in First Out). Insertion at the rear, deletion at the front.

Circular Queue : The last position is connected back to the first position to make a circle,preventing memory wastage in fixed size arrays.

Deque(Double Ended Queue) : Elements can be added or removed from both the front and the back.
